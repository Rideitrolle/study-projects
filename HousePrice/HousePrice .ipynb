### Разбор California Housing Prices

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Загрзука и обработка данных

```python
data = pd.read_csv("housing.csv") #https://www.kaggle.com/datasets/camnugent/california-housing-prices
```

```python
data.head(10)
```

```python
data.info()
```

### Очистка данных

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(data['total_bedrooms'], bins=50)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
data.dropna(subset=['total_bedrooms'],inplace = True)
```

```python
data['median_income_dollars'] = data['median_income'] * 10000
```

```python
data_code = pd.get_dummies(data, columns=['ocean_proximity'])
```

```python
data.head()
```

```python
data.info()
```

### Исследовательский анализ данных

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['housing_median_age'])
plt.title('Boxplot of Total Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['median_income'])
plt.title('Boxplot of Total Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['median_house_value'])
plt.title('Boxplot of Total Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['total_rooms'])
plt.title('Boxplot of Total Bedrooms')
plt.show()
```

```python
# Расчет корреляционной матрицы
correlation_matrix = data_code.corr()
```

### Карта корреляций

```python
plt.figure(figsize=(12, 10))

# Построение тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Настройка заголовка и отображение
plt.title('Корреляционная матрица')
plt.show()
```

```python
grouped_data = data.groupby('ocean_proximity')['median_house_value'].mean()
print(grouped_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Группировка и визуализация
plt.figure(figsize=(10, 6))
sns.barplot(x=grouped_data.index, y=grouped_data.values)
plt.title('Среднее значение median_house_value по ocean_proximity')
plt.xlabel('ocean_proximity')
plt.ylabel('Среднее значение median_house_value')
plt.xticks(rotation=45)
plt.show()
```

### Распределение числовых переменных

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Плоты распределения числовых признаков
num_cols = ['median_income', 'total_rooms', 'total_bedrooms', 'households', 'median_house_value']

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Распределение {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(data[num_cols])
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, col in enumerate(num_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(data[col])
    plt.title(f'Выбросы в {col}')
plt.tight_layout()
plt.show()

```

```python
unique_values = data['ocean_proximity'].unique()
print(unique_values)
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data[data['ocean_proximity'] == 'NEAR BAY']['median_house_value'], label='NEAR BAY')
sns.kdeplot(data[data['ocean_proximity'] == 'INLAND']['median_house_value'], label='INLAND')
sns.kdeplot(data[data['ocean_proximity'] == '<1H OCEAN']['median_house_value'], label='<1H OCEAN')
sns.kdeplot(data[data['ocean_proximity'] == 'NEAR OCEAN']['median_house_value'], label='NEAR OCEAN')
sns.kdeplot(data[data['ocean_proximity'] == 'ISLAND']['median_house_value'], label='ISLAND')
plt.title('Распределения медианной стоимости домов по категориям близости к океану')
plt.xlabel('Медианная стоимость домов')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```
